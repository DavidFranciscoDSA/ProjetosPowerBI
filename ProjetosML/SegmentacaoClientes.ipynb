Problema de Negócio:

Considerando dados históricos de clientes que realizaram compras em nossa empresa, faça o agrupamento (segmentação) dos clientes por similaridade de características em 3 grupos e envie o relatório para a equipe de Marketing.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

<h3><font color='orangered'>Carregando a base de dados</font></h3>


```python
caminho = r'C:\Users\david\Desktop\ESTUDOS\2-DSA\Microsoft Power BI Para Business Intelligence e Data Science\ProjetosPowerBI\ProjetosPowerBI\ProjetosML\dados\dados_clientes.csv'
df = pd.read_csv(caminho, sep=',')
df
```

<h3><font color='orangered'>Análise exploratória dos dados</font></h3>

```python
# Variaveis de interesse
df[['idade','renda_anual', 'pontuacao_gastos']].describe()
```

<h3><font color='orangered'>Pré-processamento dos dados</font></h3>

- Cria o padronizador dos dados
- Aplica o padronizador somente nas colunas de interesse

```python
padronizador = StandardScaler()
dados_padronizados = padronizador.fit_transform(df[['idade', 'renda_anual', 'pontuacao_gastos']])
print(dados_padronizados)
```

<h3><font color='orangered'>Construção do Modelo de Machine Learning Para Segmentação de Clientes</font></h3>

- Definimos o número de clusters (k)
- Criamos o modelo K-means
- Treinamento do modelo com os dados padronizados
- Atribuímos os rótulos dos clusters aos clientes

```python
k = 3
kmeans = KMeans(n_clusters = k)
kmeans.fit(dados_padronizados)
df['cluster'] = kmeans.labels_
df.head(150)
```

```python
# Salvamos o resultado em disco
df.to_csv('dados/segmentos2.csv', index = False)
```
